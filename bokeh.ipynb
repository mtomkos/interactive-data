```python
from bokeh.plotting import figure, show
```

```python
import pandas as pd
twitch = pd.read_csv('twitchdata-update.csv')
```

```python
twitch
```

```python
p = figure()
x = twitch['Language']
y = twitch['Followers']
p.line(x,y,line_width = 2)
p.circle(x, y,size = 20)
show(p)
```

```python
twitch_new = twitch.groupby('Language')
twitch_new.count()
```

```python
counts = twitch_new.count()
counts
```

```python
x = list(range(counts.shape[]))
y = counts['Watch time(Minutes)']

plot = figure(width=400, height=400, title='Select and Zoom',
              tools="box_select,box_zoom,lasso_select,reset")

p.line(x, y, legend_label="Temp.", line_width=2)
```

```python
x = list(range(counts.shape[0]))
y = counts['Watch time(Minutes)']

p = figure(width=1200, height=300, title="Twitch streamers by spoken language", x_axis_label="Spoken language", y_axis_label="Number of streamers")

p.line(x, y, legend_label="Number of streamers", line_width=2)

p.xaxis.ticker = list(range(counts.shape[0]))
p.xaxis.major_label_overrides = {1: 'Arabic', 2: 'Chinese', 3: 'Czech', 4: 'English', 5: 'Finnish', 6: 'French', 7: 'German', 8: 'Greek', 9: 'Hungarian', 10: 'Italian', 11: 'Japanese', 12: 'Korean', 13: 'Other', 14: 'Polish', 15: 'Portuguese', 16: 'Slovak', 17: 'Spanish', 18: 'Swedish', 19: 'Thai', 20: 'Turkish'}

show(p)
```

```python
p = figure(width=1200, height=400)

p.circle(list(range(counts.shape[0])), counts['Watch time(Minutes)'], size=20, color="navy", alpha=0.5)

# show the results
show(p)
```
