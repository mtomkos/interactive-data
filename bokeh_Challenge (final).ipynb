```python
import numpy as np
import pandas as pd
from bokeh.models import HoverTool
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, output_file
```

```python
twitch = pd.read_csv('twitchdata-update.csv')
twitch
```

```python
twitch_new = twitch.groupby('Language')
twitch_new.count()
```

```python
counts = twitch_new.count()
counts = counts.sort_values('Channel', ascending=False)
counts
```

```python
counts.index
```

```python
x = list(counts.index)
y = counts['Channel']
output_notebook()

p = figure(x_range = x, sizing_mode="stretch_width")
p.line(x, y)

show(p)
```

```python
x = list(counts.index)
y = counts['Channel']

# Add hover tool tips
hover1 = HoverTool(tooltips=[("Number of streamers", "@y")],
                  mode='vline'
                  )
# Add title, x and y labels, remove tool bar, add HoverTool feature
p2 = figure(x_range = x, sizing_mode="stretch_width", height=300, title="Twitch streamers by spoken language", 
           x_axis_label="Spoken language", y_axis_label="Number of streamers", toolbar_location=None, tools=[hover1])


# Change line color, add title for legend, change line width
p2.line(x, y, line_color="green", legend_label="Number of streamers", line_width=2)

# Add horizontal dashed lines
p2.ygrid.grid_line_alpha = 0.8
p2.ygrid.grid_line_dash = [6, 4]

show(p2)
```
