```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_html('https://www.zippia.com/user-experience-designer-jobs/demographics/')
len(data)

table_data = data[1]
table_data
```

```python
ux_data = pd.melt(
    table_data,
    id_vars='Year',
    value_vars=['Male', 'Female'],
    var_name='Gender',
    value_name='Percent'
)

ux_data['Percent'] = ux_data['Percent'].str.rstrip('%').astype('float')

ux_data = ux_data.astype({
    'Year': 'category',
    'Gender': 'category'
})

year_gender_group = ux_data.groupby(['Year', 'Gender'])
```

```python
fig_1, ax_1 = plt.subplots()

for idx, o in year_gender_group:
    if o.Gender.item() == 'Male':
        ax_1.barh(o.Year, o.Percent, left=0, label='Male', color="tab:orange")
    else:
        start_loc = 100 - o.Percent
        ax_1.barh(o.Year, o.Percent, left=start_loc, label='Female', color="tab:blue")
            
ax_1.legend(['Female', 'Male'])

ax_1.set(xlabel='% Gender', ylabel='Year', title='UX Designer Gender by Year')

plt.show()
```

```python
fig_2, ax_2 = plt.subplots()

sns.set_style('whitegrid')

sns.histplot(ux_data, y='Year', hue='Gender', weights='Percent', multiple='stack')

ax_2.set(xlabel='% Gender', ylabel='Year', title='UX Designer Gender by Year')

plt.show()
```

While navigating through this challenge, I noticed that seaborn requires much less syntax than matplotlib to create a visualization. Both graphs allowed us to stack the bars for each gender side by side, but matplotlib required an "if else" statement so the 'Female' bar overlapped the 'Male' bar.

Seaborn seems to be less time-consuming, especially when working with more detailed data considering its customizations. I would use seaborn if I'm working with extensive datasets, like keyword trends over the timespan of a year.
